```python
import tensorflow as tf
import numpy as np
from sklearn.datasets import make_blobs
```

```python
import matplotlib.pyplot as plt
```

```python
x,y=make_blobs(centers=2,random_state=42)
```

```python
x=tf.constant(x,dtype=np.float64)
y=tf.constant(y.reshape(-1,1),dtype=np.float64)
```

```python
y.shape
```

```python
plt.style.use("seaborn-talk")
```

```python
plt.scatter(x[:,0],x[:,1],c=y.numpy().flatten())
```

```python
W=tf.Variable(tf.zeros([x.shape[1],1],dtype=np.float64))
b=tf.Variable(0.,dtype=np.float64)
```

```python
leaning_rate=0.8
max_step=10
```

```python
@tf.function
def coustom_logis(x,y):
    with tf.GradientTape() as tape:
        z=tf.matmul(x,W)+b
        yhat=tf.sigmoid(z)
        error=-((y*tf.math.log(yhat)) + (1-y)*(tf.math.log(1-yhat)))
        loss=tf.reduce_mean(error)
        
    dW ,db=tape.gradient(loss ,[W,b])
    W.assign_sub(dW*leaning_rate)
    b.assign_sub(db*leaning_rate)
    
    return(loss)
```

```python
for i in range(max_step):
    print(i,coustom_logis(x,y))
    
```

```python
m,c=-W[0]/W[1],-b/W[1]
```

```python
line=np.linspace(-6,6,5)
```

```python
y_line=m*line+c
```

```python
plt.plot(y_line)
plt.scatter(x[:, 0], x[:, 1], c=y.numpy().flatten())
```
