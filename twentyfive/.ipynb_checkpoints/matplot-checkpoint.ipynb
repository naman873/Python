```python
x=np.linspace(-10,10)
plt.figure()
plt.subplot(3,1,1)
plt.plot(x,np.sin(x))

plt.subplot(3,1,2)
plt.plot(x,np.cos(x))

plt.subplot(3,2,5)
plt.plot(x,np.tan(x))

plt.subplot(3,2,6)
plt.plot(x,x**2)
plt.show()
```

```python
r = 5
theta = np.linspace(0, np.pi, 100)

X = r * np.cos(theta)
y = r * np.sin(theta)

plt.plot(X, y)
```

```python
x=np.linspace(100,20*np.pi,1000)
y=np.log(np.linspace(100,1000,1000))

z=y*np.sin(x)
q=y*np.cos(x)
```

```python
plt.plot(z,q)
plt.plot(z,z**2)
plt.plot(q,q**2)
```

```python
x=np.linspace(0,20*np.pi,1000)
y=np.log(np.linspace(100,1000,1000))

z=y*np.sin(x)
q=y*np.cos(x)
```

```python
plt.figure()

plt.plot(q , z , label="pera")
plt.plot(q, q**2 , label="per")

plt.xlabel("ye apna x ha")
plt.ylabel("ye apna y ha")
plt.title("ye graph ha")
plt.legend()
plt.show()
```
