```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.linspace(0,2*np.pi,10)
```

```python
b=np.sin(a)
```

```python
plt.plot(a,b)
#plt.scatter(a,b)
```

```python
a=np.linspace(-100,100,10)
```

```python
plt.plot(a,a**2)
plt.plot(a,a**3)
```

```python
x=np.linspace(0,2*np.pi,100)
```

```python
a=5*np.sin(x)
b=5*np.cos(x)
```

```python
plt.plot(a,b)
```

```python
y=np.linspace(0,8,100)
```

```python
t=y*np.sin(x)
u=y*np.cos(x)
```

```python
plt.plot(t,u)
```

```python
var=np.linspace(0,1000,100)
x=np.linspace(0,7*np.pi*1000,100)
a=var*np.sin(x)
b=var*np.cos(x)
```

```python
plt.plot(b,a)
```

```python
x=np.linspace(-10,10)
plt.figure()
plt.subplot(3,1,1)
plt.plot(x,np.sin(x))

plt.subplot(3,1,2)
plt.plot(x,np.cos(x))

plt.subplot(3,2,5)
plt.plot(x,np.tan(x))

plt.subplot(3,2,6)
plt.plot(x,x**2)
plt.show()
```
