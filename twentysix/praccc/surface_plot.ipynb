```python
import  numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a=np.array([1,2,3])  ## it will define number of colums
b=np.array([4,5,6,7])  ## it will define  number of rows
```

```python
a,b=np.meshgrid(a,b)
```

```python
a
```

plot surface take 3 arguments such as a and b and third one is funtion of a and b or some 
equatiion in a and b

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap="coolwarm")
plt.show()
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(a,b,a+b,cmap="rainbow")
plt.show()
```

```python
a=np.arange(-1,1,0.02)
b=a
a,b=np.meshgrid(a,b)
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap="rainbow")
plt.show()
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap="rainbow")
plt.show()
```
