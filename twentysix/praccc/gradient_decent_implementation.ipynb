```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(10)
y=(X-5)**2
```

```python
plt.style.use("seaborn")
plt.xlabel("X")
plt.ylabel("f(X)")
plt.plot(X,y)
```

```python
x=0
error=[]
lr=0.1
for i in range(50):
    grad=2*(x-5)
    x=x - lr*grad
    e=(x-5)**2
    error.append(e)
    plt.plot(X,y)
    plt.scatter(x,e)
    #print(e)
```
