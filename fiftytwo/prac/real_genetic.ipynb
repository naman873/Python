```python
virus=np.array([.8,.8])
```

```python
def small(data):
    dist=np.sum((data-virus)**2,axis=1)
    smal=dist.argsort()[:10]
    return data[smal]
```

```python
final=small(data)
```

```python
x1=random.choice(final)
```

```python
x2=random.choice(final)
```

```python
def populate(data,steps):
    
    gen=[]
    
    for i in range(steps):
        
        x1=random.choice(data)
        x2=random.choice(data)
        
        new_gen=np.array([x1[0],x2[1]])
        
        gen.append(new_gen)
        
        
    generated=np.array(gen)
    noise=np.random.randn(100,2) *(.1)
    
    return generated+noise
```

```python
new_g=populate(final,100)
```

```python
def fitness(population,virus):
    
    dist=np.sqrt(np.sum((population-virus)**2,axis=1))
    fit=dist.argsort()[:10]
    #sel=population[fit]
    return fit
```

```python
population=np.random.rand(100,2)
virus=np.array([-1.5,-1.5])

for i in range(15):
    plt.figure()
    plt.scatter(population[:,0],population[:,1])
    plt.scatter(virus[0],virus[1])
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    plt.show()
    
    index=fitness(population,virus)
    sel=population[index]
    population=populate(sel,100)
```
