```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import statsmodels.api as sm
lowess=sm.nonparametric.lowess
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(centers=3,random_state=42)
```

```python
x.shape
```

```python
y.shape
```

```python
ho=x[:,0]
```

```python
vr=x[:,1]
```

```python
plt.scatter(ho,vr,c=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(x[:,0],x[:,1],c=y)

for (m1,m2),c in zip(model.coef_,model.intercept_):
    x1=np.linspace(-10,10,100)
    x2=(-x1*m1)/m2 - (c/m2)
    plt.ylim(-10,10)
    #print("how many times loop run")
    plt.plot(x1,x2)
```

```python
color=["red","green","black","blue"]
```

```python
y_color=[color[index] for index in y  ]
```

```python
plt.scatter(x[:,0],x[:,1],c=y_color)

for index,(m1,m2),c in zip(y_color,model.coef_ , model.intercept_):
    x1=np.linspace(-10,10,100)
    x2=(-x1*m1)/m2 - (c/m2)
    plt.ylim(-10,10)
    plt.plot(x1,x2,index)
```
