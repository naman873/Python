```python
import  numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_features=2,centers=2,random_state=0)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
class coustom:
    def __init__(self,neighbour=5):
        self.neighbour=neighbour
        
    def fit(self,X,y):
        self.X=X
        self.y=y
    
    def predicted_point(self,point):
        dist=np.sum((self.X-point)**2,axis=1)
        
        min_dist=self.y[dist.argsort()[:self.neighbour]]
        items,count=np.unique(min_dist,return_counts=True)
        return items[count.argmax()]
    
    def predicted(self,X):
        result=[]
        
        for point in X:
            result.append(self.predicted_point(point))
        return result
    
    def score(self,X,y):
        yp=self.predicted(X)
        return (yp==y).mean()
```

```python
a=coustom()
```

```python
a.fit(X,y)
```

```python
a.score(X,y)
```

```python
acc=[]
for i in range(1,50):
    a=coustom(neighbour=i)
    a.fit(X,y)
    point=np.array([2,2])
    a.predicted_point(point)
    acc.append(a.score(X,y))
```

```python
plt.plot(acc)
```
