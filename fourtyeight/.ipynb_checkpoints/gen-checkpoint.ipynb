```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.preprocessing.image import load_img
```

```python
img_panda=load_img("../fifth/panda.jpg",target_size=(300,400))
img_conv=load_img("../fourtyseven/mon.jpg",target_size=(300,400))
```

```python
arr_pan=np.array(img_panda)
arr_conv=np.array(img_conv)
```

```python
arr_conv.shape,arr_pan.shape
```

```python
batch=np.array([arr_conv,arr_pan])
```

```python
batch.shape
```

```python
generator=ImageDataGenerator(rotation_range=10, horizontal_flip=True, shear_range=1.1)
```

```python
gen=generator.flow(batch,batch_size=10)
```

```python
for i in range(5):
    for gen_img in gen.next():
        plt.figure()
        plt.imshow(gen_img.astype(int))
        plt.show()
```
