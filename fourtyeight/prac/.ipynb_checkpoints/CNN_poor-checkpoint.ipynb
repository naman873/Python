```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.layers import Layer,Dense,Convolution2D,MaxPool2D,Input,Flatten
from tensorflow.keras.models import Model
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
generator=ImageDataGenerator(shear_range=2.0,horizontal_flip=True,rotation_range=20)
```

```python
gen=generator.flow_from_directory("../4classimages/data",target_size=(200,200))
```

```python
gen.class_indices
```

```python
in_layer=Input(shape=(200,200,3))
convo1=Convolution2D(filters=32,kernel_size=3)(in_layer)
pool1=MaxPool2D()(convo1)
convo2=Convolution2D(filters=32,kernel_size=3)(pool1)
pool2=MaxPool2D()(convo2)
flat=Flatten()(pool2)
d1=Dense(units=100,activation="sigmoid")(flat)
d2=Dense(units=10,activation="sigmoid")(d1)
out=Dense(units=4,activation="softmax")(d2)
```

```python
model=Model(inputs=in_layer,outputs=out)
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit_generator(gen,epochs=2)
```

```python
img=load_img("dog.jpg",target_size=(200,200))

np_img=np.array(img)

im=np_img.reshape(-1,200,200,3)

model.predict(im).argmax()
```

```python
m2=Model(inputs=in_layer,outputs=model.layers[3].output)
```

```python
out=m2.predict(gen)
```
