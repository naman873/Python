```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator


from tensorflow.keras.utils import to_categorical
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
generator=ImageDataGenerator(rotation_range=10, shear_range=1.2,horizontal_flip=True)
```

```python
gen=generator.flow_from_directory("4classimages/data/")
```

```python
gen.class_indices
```

```python
x,y=gen.next()
```

```python
x.shape,y.shape
```

```python
for gen_img,indices in zip(x,y):
        plt.figure()
        plt.title(str(indices))
        plt.imshow(gen_img.astype(int))
        plt.show()
```
