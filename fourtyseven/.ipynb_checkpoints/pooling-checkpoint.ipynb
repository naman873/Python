```python
import numpy as np
import matplotlib.pyplot as plt
import PIL 
```

```python
img=PIL.Image.open("mon.jpg")
```

```python
small=img.resize(size=(400,300))
```

```python
img=np.array(small).mean(axis=2)
```

```python
plt.imshow(img,cmap="gray")
```

```python
result=np.zeros([img.shape[0]//2,img.shape[1]//2])
for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut=img[2*row:2*row+2,2*col:2*col+2]
        result[row,col]=np.max(cut)
```

```python
plt.imshow(result,cmap="gray")
```

```python
result1=np.zeros([img.shape[0]//2,img.shape[1]//2])

for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut=img[2*row:2*row+2,2*col:2*col+2]
        result1[row,col]=np.mean(cut)
```

```python
plt.imshow(result,cmap="gray")
```
