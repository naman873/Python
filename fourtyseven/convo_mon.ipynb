```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img=Image.open("mon.jpg")
```

```python
img.size
```

```python
small=img.resize(size=(400,300))  ### y,x
```

```python
small
```

```python
small.size
```

```python
img=np.array(small).mean(axis=2)
```

```python
plt.imshow(img,cmap="gray")
```

```python
img.shape
```

```python
padded_img=np.pad(array=img,pad_width=1)
```

```python
padded_img.shape
```

```python
result=np.zeros_like(img)
```

```python
result.shape
```

```python
fill=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
```

```python
for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut=padded_img[row:row+3,col:col+3]
        result[row,col]=(cut*fill).mean()
```

```python
plt.imshow(result,cmap="gray")
```

```python
result1=np.zeros_like(img)
```

```python
fill1=np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
```

```python
for row in range(result1.shape[0]):
    for col in range(result1.shape[1]):
        cut=padded_img[row:row+3,col:col+3]
        result1[row,col]=(cut*fill1).mean()
```

```python
plt.imshow(result1,cmap="gray")
```

```python
result2=np.zeros_like(img)
```

```python
fill2=np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
```

```python
for row in range(result2.shape[0]):
    for col in range(result2.shape[1]):
        cut=padded_img[row:row+3,col:col+3]
        result2[row,col]=abs((cut*fill2).mean())
```

```python
plt.imshow(result2,cmap="gray")
```
