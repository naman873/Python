```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img=Image.open("../horse.jpg")
```

```python
img
```

```python
small=img.resize((400,300))   ## y,x
```

```python
small
```

```python
num_img=np.array(small)
```

```python
num_img.shape
```

```python
num_img_b=num_img.mean(axis=2)
```

```python
num_img_b.shape
```

```python
plt.imshow(num_img_b,cmap="gray")
```

```python
padded_img=np.pad(num_img_b,pad_width=1,mode="constant")
```

```python
plt.imshow(padded_img,cmap="gray")
```

```python
fill=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])  ## horizontal
```

```python
final=np.zeros_like(num_img_b)
```

```python
final.shape
```

```python
for x in range(final.shape[0]):
    for y in range(final.shape[1]):
        cut=padded_img[x:x+3,y:y+3]
        final[x,y]=np.mean(fill*cut)
```

```python
plt.imshow(final,cmap="gray")
```

```python
fill1=np.array([[1,0,-1],[1,0,-1],[1,0,-1]])  
```

```python
final1=np.zeros_like(num_img_b)
```

```python
final1.shape
```

```python
for x in range(final1.shape[0]):
    for y in range(final1.shape[1]):
        cut=padded_img[x:x+3,y:y+3]
        final1[x,y]=np.mean(cut*fill1)
```

```python
plt.imshow(final1,cmap="gray")
```

```python
for x in range(final1.shape[0]):
    for y in range(final1.shape[1]):
        cut=padded_img[x:x+3,y:y+3]
        final1[x,y]=abs(np.mean(cut*fill1))
```

```python
plt.imshow(final1,cmap="gray")
```

```python
img=Image.open("../mon.jpg")
```

```python
img
```

```python
small=img.resize((400,300))
```

```python
small
```

```python
num_img=np.array(small)
```

```python
num_img_b=num_img.mean(axis=2)
```

```python
num_img.shape
```

```python
plt.imshow(num_img_b,cmap="gray")
```

```python
padded_img=np.pad(num_img_b,mode="constant",pad_width=1)
```

```python
fill=np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
```

```python
final3=np.zeros_like(num_img_b)
```

```python
for i in range(final3.shape[0]):
    for j in range(final3.shape[1]):
        cut=padded_img[i:i+3,j:j+3]
        final3[i,j]=abs(np.mean(cut*fill))
```

```python
plt.imshow(final3,cmap="gray")
```
