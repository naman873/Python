```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img=Image.open("../horse.jpg")
```

```python
small=img.resize((400,300))
```

```python
num_img=np.array(small)
```

```python
num_img_b=num_img.mean(axis=2)
```

```python
pad_img=np.pad(num_img_b,mode="constant",pad_width=1)
```

```python
plt.imshow(pad_img)
```

```python
final=np.zeros([num_img.shape[0]//2,num_img.shape[1]//2])
```

```python
for x in range(final.shape[0]):
    for y in range(final.shape[1]):
        cut=pad_img[2*x:2*x+2,2*y:2*y+2]
        final[x,y]=np.max(cut)
```

```python
plt.imshow(final,cmap="gray")
```
