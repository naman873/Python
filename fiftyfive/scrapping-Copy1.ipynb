```python
url="https://www.goodreads.com/quotes/tag/{}?page={}"
```

```python
def scrap(url):
    data = requests.get(url)

    soup = BeautifulSoup(data.text)

    quote_divs = soup.find_all("div", attrs={"class":"quote"})

    cont = []

    for quo in quote_divs:
        total=quo.find("div",attrs={"class":"quoteText"})
        text=total.text
    #print(text)
        quote=text.strip().split("\n")[0]
        author=total.find("span").text.strip()
        left = quo.find("div", attrs={"class":"left"})
        tags = ", ".join([a.text for a in left.find_all("a")])
        cont.append([quote,author,tags])
    
    return  cont
```

```python
emotion = input("Enter the emotion : ")
pages = int(input("Enter number of pages : "))
```

```python
content=[]
for num in range(1,pages+1):
    content.extend(scrap(url.format(emotion,num)))
    print("processed",num)
```
