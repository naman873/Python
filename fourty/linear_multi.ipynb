```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
import numpy as np
import tensorflow as tf
```

```python
x,y=make_regression(n_features=5,random_state=42)
```

```python
x=tf.constant(x,dtype=np.float64)
y=tf.constant(y.reshape(-1,1),dtype=np.float64)
```

```python
ones=tf.ones((x.shape[0],1),dtype=np.float64)
x_mod=tf.concat([ones,x],axis=1)
```

```python
theta=tf.Variable(np.zeros([x_mod.shape[1],1]),dtype=np.float64)
```

```python
#training info
train_steps = 30
learning_rate = 0.01
```

```python
@tf.function
def func(x,y):
    with tf.GradientTape() as tape:
        yhat=tf.matmul(x,theta)
       #tp=tf.matmul(x_mod,theta)
        
        error=yhat-y
        loss=tf.reduce_mean(error**2)
    
    dt=tape.gradient(loss ,theta)
    theta.assign_sub(dt*learning_rate)
    return loss
```

```python
for i in range(train_steps):
    print(func(x_mod,y)) 
```
