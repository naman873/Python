```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

```python
img=Image.open("drS.jpg")
```

```python
np_img=np.array(img)
```

```python
np_img.shape
```

```python
plt.imshow(np_img)
```

```python
np_img.dtype
```

```python
np_img.shape
```

```python
X=np.reshape(np_img,(-1,3))
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(8)
```

```python
model.fit(X)
```

```python
pallet=model.cluster_centers_.astype(np.uint8)
```

```python
pallet
```

```python
model.labels_
```

```python
plt.imshow(pallet.reshape(1, -1, 3))
```

```python
output=np.array([pallet[klass] for klass in model.labels_])
```

```python
output.shape
```

```python
output_img=output.reshape(np_img.shape)
```

```python
plt.imshow(output_img)
```
