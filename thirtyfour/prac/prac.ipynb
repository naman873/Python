```python
import numpy as np
from PIL import  Image
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
image=Image.open("drS.jpg")
```

```python
image
```

```python
np_img=np.array(image)
```

```python
np_img.dtype
```

```python
np_img.shape
```

```python
plt.imshow(np_img)
```

```python
X=np.reshape(np_img,(-1,3))
```

```python
X.shape
```

```python
model=KMeans()
```

```python
model.fit(X)
```

```python
pallets=model.cluster_centers_.astype(int)
```

```python
pallets
```

```python
labels=model.labels_
```

```python
labels[270:290]
```

```python
pallets.shape
```

```python
np.unique(labels)
```

```python
pallets=np.array(pallets,dtype=np.uint8)
```

```python
pallets
```

```python
plt.imshow(pallets.reshape(1,-1,3))
```

```python
output=np.array([pallets[klass] for klass in model.labels_])
```

```python
output.shape
```

```python
out_img=output.reshape(np_img.shape)
```

```python
plt.imshow(out_img)
```
