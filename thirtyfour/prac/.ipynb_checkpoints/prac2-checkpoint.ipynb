```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from PIL import Image
```

```python
image=Image.open("dr_S2.jpg")
```

```python
image
```

```python
np_img=np.array(image)
```

```python
plt.imshow(np_img)
```

```python
np_img.dtype
```

```python
np_img.shape
```

```python
X=np.reshape(np_img,(-1,3))
```

```python
X.shape
```

```python
model=KMeans(16)
```

```python
model.fit(X)
```

```python
pallet=model.cluster_centers_.astype(int)
```

```python
plt.imshow(pallet.reshape(1,-1,3))
```

```python
pallet
```

```python
#pallet=np.array(pallet,dtype=np.int8)
```

```python
model.labels_
```

```python
output=np.array([pallet[klass] for klass in model.labels_])
```

```python
output.shape
```

```python
out_img=output.reshape(np_img.shape)
```

```python
plt.imshow(out_img)
```
