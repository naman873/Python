```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
image=Image.open("bhai.jpg")
```

```python
plt.imshow(image)
```

```python
np_img=np.array(image)
```

```python
#plt.imshow(np_img)
```

```python
np_img.shape
```

```python
X=np_img.reshape(-1,3)
```

```python
X.shape
```

```python
model=KMeans(n_clusters=8)
```

```python
model.fit(X)
```

```python
model.labels_
```

```python
pallet=model.cluster_centers_.astype(np.uint8)
```

```python
plt.imshow(pallet.reshape(1,-1,3))
```

```python
output=np.array([pallet[klass] for klass in model.labels_])
```

```python
output.shape
```

```python
output_img=output.reshape(np_img.shape)
```

```python
plt.imshow(output_img)
```
